```python
class McCullochPittsNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def activate(self, inputs):
        # Calculate weighted sum
        weighted_sum = sum(w * x for w, x in zip(self.weights, inputs))
        # Activation function
        return 1 if weighted_sum >= self.threshold else 0

# Logic gates using McCulloch-Pitts model
def AND_gate(x1, x2):
    neuron = McCullochPittsNeuron(weights=[1, 1], threshold=1.5)
    return neuron.activate([x1, x2])

def OR_gate(x1, x2):
    neuron = McCullochPittsNeuron(weights=[1, 1], threshold=0.5)
    return neuron.activate([x1, x2])

def NOT_gate(x):
    neuron = McCullochPittsNeuron(weights=[-1], threshold=-0.5)
    return neuron.activate([x])

# Test the gates
print("AND Gate Outputs:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1} AND {x2} = {AND_gate(x1, x2)}")

print("\nOR Gate Outputs:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1} OR {x2} = {OR_gate(x1, x2)}")

print("\nNOT Gate Outputs:")
for x in [0, 1]:
    print(f"NOT {x} = {NOT_gate(x)}")
```

```python

import matplotlib.pyplot as plt

def plot_logic_gate(gate_func, title):
  x1_values = [0, 1]
  x2_values = [0, 1]
  outputs = [[gate_func(x1, x2) for x2 in x2_values] for x1 in x1_values]

  fig, ax = plt.subplots()
  ax.set_title(title)
  ax.set_xlabel("x1")
  ax.set_ylabel("x2")

  for x1 in x1_values:
      for x2 in x2_values:
          if gate_func(x1, x2) == 1:
              ax.scatter(x1, x2, color='blue', marker='o', s=100)
          else:
              ax.scatter(x1, x2, color='red', marker='x', s=100)

  plt.show()


plot_logic_gate(AND_gate, "AND Gate")


plot_logic_gate(OR_gate, "OR Gate")

def plot_not_gate():
  x_values = [0, 1]
  outputs = [NOT_gate(x) for x in x_values]

  fig, ax = plt.subplots()
  ax.set_title("NOT Gate")
  ax.set_xlabel("x")
  ax.set_ylabel("Output")

  for x, output in zip(x_values, outputs):
      if output == 1:
          ax.scatter(x, output, color='blue', marker='o', s=100)
      else:
          ax.scatter(x, output, color='red', marker='x', s=100)

  plt.show()

plot_not_gate()
```
